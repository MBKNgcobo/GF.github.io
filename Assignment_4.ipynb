##Importing Libraries

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
import os
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from great_tables import md, html
import great_tables as gt
import numpy as np
import re
import skimpy
import plotnine as p9
%matplotlib inline
from plotnine import (
ggplot,
aes,
after_stat,
stage,
geom_bar,
geom_line,
geom_point,
geom_boxplot,
geom_area,
geom_histogram,
geom_errorbar,
geom_text,
geom_label,
coord_flip
 )


import plotnine as p9
p9.theme_set(p9.theme_minimal())
```

```python
#Importing
df = pd.read_csv(filepath_or_buffer='/content/Subsistence Retail Consumer Data - Sheet1.csv - Sheet1.csv')
offending_rows= df[df.notna()]
print(offending_rows)
df = df[['PS2', 'CT1', 'PV1', 'PPQ1', 'PI3']].dropna()
sns.set(style='whitegrid')
skimpy.skim(df)
```

##Regression Analysis

```python
cols = ['CT1', 'PS2', 'PV1', 'PPQ1', 'PI3']
regression_data = df[cols].dropna()


X = regression_data[['CT1', 'PS2', 'PV1', 'PPQ1']]
y = regression_data['PI3']

# Add constant (intercept)
X = sm.add_constant(X)

model = sm.OLS(y, X).fit()

print(model.summary())
```

##Interpretation of Regression Analysis
PV1 (Perceived Value) and PPQ1 (Product Quality) are the strongest drivers of Purchase Intention.

PS2 (possibly Perceived Sustainability or Satisfaction) has a modest but significant positive impact.

CT1 (Customer Trust or Expectation match) is not statistically significant at the 5% level but is borderline.

##PS2: Willingness to pay higher price for store proximity

```python
# Boxplot of Purchase Intention by PS2
plt.figure(figsize=(6,4))
sns.boxplot(x='PS2', y='PI3', data=df, palette='crest')
plt.title('Purchase Intention vs PS2 (Proximity & Price Sensitivity)')
plt.xlabel('PS2: Willingness to pay more for store proximity')
plt.ylabel('Purchase Intention (PI)')
plt.show()

# Correlation between PS2 and PI3
corr_ps2 = df['PS2'].corr(df['PI3'])
print(f'Correlation (PS2 vs PI3): {corr_ps2:.2f}')
```

```python
funnel_data = df.groupby('PS2')['PI3'].mean().sort_index(ascending=False).reset_index()

plt.figure(figsize=(8, 6))
plt.barh(funnel_data['PS2'].astype(str), funnel_data['PI3'], color='skyblue')
plt.xlabel('Average Purchase Intention (PI3)')
plt.ylabel('PS2 Level')
plt.title('Funnel-style Chart: PS2 vs Average Purchase Intention (PI3)')
plt.gca().invert_yaxis()  # To simulate funnel direction (top to bottom)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

##Analysis of Price Sensitivity on Purchasing Intention.
As PS2 increases, the median and overall distribution of PI also increase.
This suggests that people more willing to pay for store proximity tend to have higher purchase intentions..This plot supports the idea that greater willingness to pay for proximity correlates with higher purchase intention, which could be valuable for retail location strategies and pricing models.

##CT1: Store meets your expectations

```python
# Boxplot of Purchase Intention by CT1
plt.figure(figsize=(6,4))
sns.boxplot(x='CT1', y='PI3', data=df, palette='flare')
plt.title('Purchase Intention vs CT1 (The grocery store always meets my expectations)')
plt.xlabel('CT1: Store meets customer expectations')
plt.ylabel('Purchase Intention (PI)')
plt.show()

# Correlation between CT1 and PI3
corr_ps2 = df['CT1'].corr(df['PI3'])
print(f'Correlation (CT1 vs PI3): {corr_ps2:.2f}')
```

##Analysis of CT1 on Purchasing Intention
There is a slightly increasing trend in purchase intention as CT1 increases.
Customers who strongly agree (CT1 = 5) that the store meets expectations tend to show higher purchase intention.
CT1 = 1 and CT1 = 5 show wider spread and more outliers, suggesting diverse customer opinions.
CT1 = 2 has a tighter distribution, indicating more consistent perceptions among those respondents.
While not as strong as the product quality (PPQ1) effect seen earlier, there is still a positive relationship between customers’ expectations being met and their purchase intention. This supports the idea that customer satisfaction plays a meaningful role in driving intent to buy.

##Correlation Heatmap: Predictors vs Purchase intention

```python
corr_vars = ['PS2', 'CT1', 'PV1', 'PPQ1', 'PI3']
df_corr = df[corr_vars].dropna()

# Compute correlation matrix
corr_matrix = df_corr.corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap: Predictors vs Purchase Intention")
plt.show()
```

##Analysis for Correlation Heatmap
This correlation heatmap shows how different predictors (survey variables) are statistically associated with Purchase Intention (PI3) and with each other. The values in each cell represent the Pearson correlation coefficient (range: -1 to +1), where:

1. 1.0 indicates a perfect positive correlation.

2. 0 indicates no correlation.

3. -1.0 indicates a perfect negative correlation.

##Results
PPQ1 (0.51): Strongest correlation with purchase intention. Product quality is a key driver.

PV1 (0.49): Perceived value is also strongly correlated.

PS2 (0.36): Moderate correlation.

CT1 (0.24): Lowest correlation among predictors.

There's a general pattern of moderate to strong positive correlations among all predictors, indicating interconnected consumer perceptions.

##PPQ1: Overall product quality is good

```python
# Boxplot of Purchase Intention by PPQ1
plt.figure(figsize=(6,4))
sns.boxplot(x='PPQ1', y='PI3', data=df, palette='pastel')
plt.title('Purchase Intention vs PPQ1 (Product Quality)')
plt.xlabel('PPQ1: Overall quality of products is good')
plt.ylabel('Purchase Intention (PI)')
plt.show()

# Correlation between PPQ1 and PI
corr_ppq1 = df['PPQ1'].corr(df['PI3'])
print(f'Correlation (PPQ1 vs PI): {corr_ppq1:.2f}')
```

##Analysis for Bar Graph
There is a clear upward trend: as purchase intention increases, the average perceived product quality also increases.

Customers who rated PI3 as 5 also rated PPQ1 highest (~4.3 – 4.4).

Those who gave the lowest purchase intention (PI3 = 1) rated product quality the lowest (~2.2).The growth appears progressive and consistent, with each step up in PI3 linked to a higher average PPQ1.

This suggests a strong positive relationship between perceived product quality and purchase intention. As customers perceive the product to be of higher quality, they are more inclined to intend to purchase it.
